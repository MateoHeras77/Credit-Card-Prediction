```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
appl = pd.read_csv("../Credit-Card-Prediction/application_record.csv", sep=",")
balance = pd.read_csv("./credit_record.csv", sep=",")
```

```python
print(appl.shape)
appl = appl.drop_duplicates(subset=['CODE_GENDER','FLAG_OWN_CAR','FLAG_OWN_REALTY','CNT_CHILDREN','AMT_INCOME_TOTAL','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE','DAYS_BIRTH','DAYS_EMPLOYED','FLAG_MOBIL','OCCUPATION_TYPE'],
                    keep='first')
print(appl.shape)
```

```python
balance_v2 = balance.copy()
balance[balance["ID"]==5150482]
```

```python
Ventana = []
a = 5001711
i = -1
for row in balance['ID']:    
    if row == a:
        i += 1
        Ventana.append(i)
    else:
        a = row 
        i = 0
        Ventana.append(i)
balance_v2["Ventana"]=Ventana
```

```python
balance_v2 = balance_v2.pivot(values='STATUS',index='ID',columns='Ventana')
balance_v2['ID'] = balance_v2.index
balance_v2.reset_index(drop = True, inplace = True)
balance_v2
```

# Eliminacion de Datos NaN

```python
data = balance_v2.iloc[:,0:13]
data['ID'] = balance_v2['ID']
data
```

```python
data_v2 = data.dropna(axis=0)
print (data_v2.dropna(axis=0).shape)
print (data_v2.dropna(axis=0).isnull().sum())
data_v2
```

```python
data_v2.set_index("ID",inplace=True)
data_v2[1].unique()
```

```python
data_v2[50:100]
```

```python
data_v3= data_v2.replace(["X","C"],[-1,-2]).astype(int)
print (data_v3.dtypes)
data_v3
```

# AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering

agglomerative = AgglomerativeClustering(n_clusters=2)

data_v3_agglomerative = data_v3.copy()

yhat= agglomerative.fit_predict(data_v3_agglomerative)
```

```python
data_v3_agglomerative["lab_AggClu"]=yhat
```

# BIRCH

```python
import numpy as np
from matplotlib import pyplot
from sklearn.cluster import Birch

#data_v3_BIRCH = data_v3.copy()

model = Birch(threshold=0.01, n_clusters=2)
model.fit(data_v3)

yhat2 = model.predict(data_v3)

data_v3_agglomerative['lab_BIRCH'] = yhat2
```

```python
f,axes =plt.subplots(5,3,figsize=(30,24))

sns.countplot(x=0, hue='labels',ax=axes[0,0], data=data_v3_BIRCH)
sns.countplot(x=1, hue='labels',ax=axes[0,1], data=data_v3_BIRCH)
sns.countplot(x=2, hue='labels',ax=axes[0,2], data=data_v3_BIRCH)
sns.countplot(x=3, hue='labels',ax=axes[1,0], data=data_v3_BIRCH)
sns.countplot(x=4, hue='labels',ax=axes[1,1], data=data_v3_BIRCH)
sns.countplot(x=5, hue='labels',ax=axes[1,2], data=data_v3_BIRCH)
sns.countplot(x=6, hue='labels',ax=axes[2,0], data=data_v3_BIRCH)
sns.countplot(x=7, hue='labels',ax=axes[2,1], data=data_v3_BIRCH)
sns.countplot(x=8, hue='labels',ax=axes[2,2], data=data_v3_BIRCH)
sns.countplot(x=9, hue='labels',ax=axes[3,0], data=data_v3_BIRCH)
sns.countplot(x=10, hue='labels',ax=axes[3,1], data=data_v3_BIRCH)
sns.countplot(x=11, hue='labels',ax=axes[3,2], data=data_v3_BIRCH)
sns.countplot(x=12, hue='labels',ax=axes[4,0], data=data_v3_BIRCH)
```

# K Means

```python
from sklearn.cluster import KMeans

#data_v3_Kmeans = data_v3.copy()

model = KMeans(n_clusters=2)
model.fit(data_v3)

yhat3 = model.predict(data_v3) 

data_v3_agglomerative['lab_KMean'] = yhat3
```

```python
f,axes =plt.subplots(5,3,figsize=(30,24))

sns.countplot(x=0, hue='labels',ax=axes[0,0], data=data_v3_Kmeans)
sns.countplot(x=1, hue='labels',ax=axes[0,1], data=data_v3_Kmeans)
sns.countplot(x=2, hue='labels',ax=axes[0,2], data=data_v3_Kmeans)
sns.countplot(x=3, hue='labels',ax=axes[1,0], data=data_v3_Kmeans)
sns.countplot(x=4, hue='labels',ax=axes[1,1], data=data_v3_Kmeans)
sns.countplot(x=5, hue='labels',ax=axes[1,2], data=data_v3_Kmeans)
sns.countplot(x=6, hue='labels',ax=axes[2,0], data=data_v3_Kmeans)
sns.countplot(x=7, hue='labels',ax=axes[2,1], data=data_v3_Kmeans)
sns.countplot(x=8, hue='labels',ax=axes[2,2], data=data_v3_Kmeans)
sns.countplot(x=9, hue='labels',ax=axes[3,0], data=data_v3_Kmeans)
sns.countplot(x=10, hue='labels',ax=axes[3,1], data=data_v3_Kmeans)
sns.countplot(x=11, hue='labels',ax=axes[3,2], data=data_v3_Kmeans)
sns.countplot(x=12, hue='labels',ax=axes[4,0], data=data_v3_Kmeans)
```

# Gaussian Mixture Model

```python
data_v3_GMM = data_v3.copy()

from sklearn.mixture import GaussianMixture

model = GaussianMixture(n_components=2)
yhat4 = model.fit_predict(data_v3)

data_v3_agglomerative['lab_GMM'] = yhat4
```

```python
f,axes =plt.subplots(5,3,figsize=(30,24))

sns.countplot(x=0, hue='labels',ax=axes[0,0], data=data_v3_GMM)
sns.countplot(x=1, hue='labels',ax=axes[0,1], data=data_v3_GMM)
sns.countplot(x=2, hue='labels',ax=axes[0,2], data=data_v3_GMM)
sns.countplot(x=3, hue='labels',ax=axes[1,0], data=data_v3_GMM)
sns.countplot(x=4, hue='labels',ax=axes[1,1], data=data_v3_GMM)
sns.countplot(x=5, hue='labels',ax=axes[1,2], data=data_v3_GMM)
sns.countplot(x=6, hue='labels',ax=axes[2,0], data=data_v3_GMM)
sns.countplot(x=7, hue='labels',ax=axes[2,1], data=data_v3_GMM)
sns.countplot(x=8, hue='labels',ax=axes[2,2], data=data_v3_GMM)
sns.countplot(x=9, hue='labels',ax=axes[3,0], data=data_v3_GMM)
sns.countplot(x=10, hue='labels',ax=axes[3,1], data=data_v3_GMM)
sns.countplot(x=11, hue='labels',ax=axes[3,2], data=data_v3_GMM)
sns.countplot(x=12, hue='labels',ax=axes[4,0], data=data_v3_GMM)
```
